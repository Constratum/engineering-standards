### Terrain/height Multiplier ($M_zcat$)

```python
Table4_1 = pd.DataFrame([
    [3, 0.97, 0.91, 0.87, 0.83, 0.75],
    [5, 1.01, 0.91, 0.87, 0.83, 0.75],
    [10, 1.08, 1.00, 0.92, 0.83, 0.75],
    [15, 1.12, 1.05, 0.97, 0.89, 0.75],
    [20, 1.14, 1.08, 1.01, 0.94, 0.75],
    [30, 1.18, 1.12, 1.06, 1, 0.8],
    [40, 1.21, 1.16, 1.10, 1.04, 0.85],
    [50, 1.23, 1.18, 1.13, 1.07, 0.9],
    [75, 1.27, 1.22, 1.17, 1.12, 0.98],
    [100, 1.31, 1.24, 1.20, 1.16, 1.03],
    [150, 1.36, 1.27, 1.24, 1.21, 1.11],
    [200, 1.39, 1.29, 1.27, 1.24, 1.16]],
    columns = ["Height", "TC1", "TC2", "TC2.5", "TC3", "TC4"]
)

Table4_1
```

```python
def interpolation(height):
    indices = Table4_1['Height'].searchsorted([height], side='right')
    lower_bound_index = indices[0] - 1
    upper_bound_index = indices[0] if indices[0] != len(Table4_1) else indices[0] - 1
    
    # Handling the boundary condition where height is less than the smallest value in the dataframe
    if lower_bound_index == -1:
        lower_bound_index = 0
        upper_bound_index = 0

    height_low = Table4_1['Height'][lower_bound_index]
    height_high = Table4_1['Height'][upper_bound_index]

    # Check for division by zero
    if height_high == height_low:
        interpolation_hn = 0
    else:
        interpolation_hn = (height - height_low) / (height_high - height_low)

    return interpolation_hn, lower_bound_index, upper_bound_index

def calc_mz_cat(height, Terrain_category):
    interpolation_height, lower_bound_index, upper_bound_index = interpolation(height)
    mz_cat_low = Table4_1[Terrain_category][lower_bound_index]
    mz_cat_high = Table4_1[Terrain_category][upper_bound_index]
    Mz_cat = mz_cat_low + interpolation_height * (mz_cat_high - mz_cat_low)
    
    return Mz_cat
```

### $M_s$ Shielding multiplier

```python
# Sheilding_parameters

def calc_s_sheilding_par(hs, bs, h, ns):
    ls = h * (10 / ns + 5)
    s = ls / ((hs * bs) ** 0.5)
    
    return s

Table4_2 = pd.DataFrame({
    "sheilding_parameter": [1.5, 3.0, 6.0, 12],
    "sheilding_multiplier": [0.7, 0.8, 0.9, 1.0]},
    index = [1, 2, 3, 4]
)

def interpolation_height_table4_2(s):
    indices = Table4_2['sheilding_parameter'].searchsorted([s], side='right')
    lower_bound_index = indices[0] - 1
    upper_bound_index = indices[0] if indices[0] != len(Table4_2) else indices[0] - 1
    height_low = Table4_2['sheilding_parameter'][lower_bound_index]
    height_high = Table4_2['sheilding_parameter'][upper_bound_index]
    interpolation_hn = (s - height_low) / (height_high - height_low)
    return interpolation_hn, lower_bound_index, upper_bound_index

def calc_ms(s):
    if s >= 12:
        Ms = 1.0
    elif s <= 1.5:
        Ms = 0.7
    else:
        interpolation_height, lower_bound_index, upper_bound_index = interpolation_height_table4_2(s)
        ms_low = Table4_2.loc[lower_bound_index, "sheilding_multiplier"]
        ms_high = Table4_2.loc[upper_bound_index, "sheilding_multiplier"]
        Ms = ms_low + interpolation_height * (ms_high - ms_low)
        
    return Ms


M_S= calc_ms(0.5)
M_S
```

### $(M_lee)$ Lee multiplier

```python
# Topographic multiplier (Mt)

def calc_mh(H, lu, x, z, l1, l2, zone):
    
    if H / (2 * lu) < 0.05:
        Mh = 1.0
    elif H / (2 * lu) >= 0.5 and H / (2 * lu) <= 0.45:
        Mh = 1.0 + (H / 3.5 * (z + l1)) * (1 - (np.abs(x) / l2))
        
    else:
        if zone == "rectangular":
            Mh = 1.0 + 0.71 * (1 - np.abs(x) / l2)
        else:
            Mh = 1.0 + (H / 3.5 * (z + l1)) * (1 - (np.abs(x) / l2))
    
    return Mh


## Mt

def calc_mt(Mh, Mlee, E):
    return Mh * Mlee * (1 + 0.00015 * E)
```

```python
def site_wind_speed(p, location, height, Terrain_category):
    
    Vr = calc_vr(p, location)
    Mz_cat_value = calc_mz_cat(height, Terrain_category)
    
    return Vr * Md * (Mz_cat_value * Ms * Mt)
```
